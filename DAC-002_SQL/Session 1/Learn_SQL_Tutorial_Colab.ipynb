## üß© Setting Up a Simple Database
Let's create a small **Driver Database** with two tables: `driver` and `team`.

```python
%%sql

DROP TABLE IF EXISTS drivers;
DROP TABLE IF EXISTS team;

CREATE TABLE drivers (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    races  INTEGER,
    team_id INTEGER
);

CREATE TABLE team (
    id INTEGER PRIMARY KEY,
    team_name TEXT,
    team_principal TEXT
);

INSERT INTO team VALUES
(1, 'Red Bull', 'Laurent Mekies'),
(2, 'Ferrari', 'Fred Vasseur'),
(3, 'McLaren', 'Andrea Stella');

INSERT INTO drivers VALUES
(1, 'Max Verstappen', 28, 227, 1),
(3, 'Daniel Ricciardo', 36, 257, 1),
(16, 'Charles Leclerc', 27, 165, 2),
(44, 'Lewis Hamilton', 40, 374, 2),
(81, 'Oscar Piastri', 24, 64, 3);
```

## üëÄ SELECT Basics
`SELECT` is how we ask questions in SQL. Think of it as saying: *‚ÄúHey database, show me this info!‚Äù*

```python
%%sql
# Show the drivers table

SELECT * FROM drivers;
```

```python
%%sql
# Show the team table

SELECT * FROM team;
```

## üéØ Filtering Data with WHERE
Use `WHERE` to filter results ‚Äî it‚Äôs like setting conditions.

```python
%%sql
-- Drivers with more than 150 race starts
SELECT name, races FROM drivers WHERE races > 150;
```

### üîç LIKE and Wildcards
You can use `LIKE` to find text patterns ‚Äî `%` means 'anything'. For example, find drivers with names starting with 'M':

```python
%%sql
# Select drivers starting with 'M'

SELECT * FROM drivers
WHERE name LIKE 'M%';
```

## üìä Sorting and Limiting Results
Use `ORDER BY` to sort, and `LIMIT` to control how many rows you see.

```python
%%sql
SELECT name, races
FROM drivers
ORDER BY races
DESC LIMIT 3;
```

## üßÆ Aggregates and GROUP BY
SQL can summarize data using functions like `COUNT`, `AVG`, `SUM`, etc.

```python
%%sql
-- Average race starts by team
SELECT team_id,
AVG(races) AS avg_races
FROM drivers
GROUP BY team_id;
```

```python
%%sql

SELECT team_id,
AVG(age) AS avg_age
FROM drivers
GROUP BY team_id;
```

## ü§ù Understanding JOINs
JOINs connect related tables together. For example, combine `drivers` and `teams` to see each driver's team name.

```python
%%sql
# See each drivers team name

SELECT d.name, t.team_name
FROM drivers d
INNER JOIN team t ON d.team_id = t.id;
```

## üß© Subqueries ‚Äî Queries inside Queries
You can nest a query inside another. Example: Find drivers who've raced above the average number.

```python
%%sql

SELECT name, races
FROM drivers
WHERE races > (SELECT AVG(races) FROM drivers);
```

## üí° Common SQL Mistakes and Tips
Here are some common beginner mistakes (and how to fix them):

1. ‚ùå **Forgetting WHERE in DELETE/UPDATE** ‚Üí Always check before you run it!
2. ‚ùå **Using = NULL** ‚Üí Use `IS NULL` or `IS NOT NULL` instead.
3. ‚ùå **Mixing up INNER vs LEFT JOIN** ‚Üí INNER shows matches only; LEFT keeps all from the left table.
4. ‚ùå **GROUP BY without aggregates** ‚Üí Only use `GROUP BY` with functions like `COUNT`, `AVG`, etc.
5. ‚ùå **Forgetting quotes** ‚Üí Strings go in single quotes `'like this'`.
6. üí° **Pro Tip:** When exploring a new table, use `LIMIT 5` to avoid huge outputs!
