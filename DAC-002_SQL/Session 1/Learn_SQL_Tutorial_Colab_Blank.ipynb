## üí° Common SQL Mistakes and Tips
Here are some common beginner mistakes (and how to fix them):

1. ‚ùå **Forgetting WHERE in DELETE/UPDATE** ‚Üí Always check before you run it!
2. ‚ùå **Using = NULL** ‚Üí Use `IS NULL` or `IS NOT NULL` instead.
3. ‚ùå **Mixing up INNER vs LEFT JOIN** ‚Üí INNER shows matches only; LEFT keeps all from the left table.
4. ‚ùå **GROUP BY without aggregates** ‚Üí Only use `GROUP BY` with functions like `COUNT`, `AVG`, etc.
5. ‚ùå **Forgetting quotes** ‚Üí Strings go in single quotes `'like this'`.
6. üí° **Pro Tip:** When exploring a new table, use `LIMIT 5` to avoid huge outputs!
